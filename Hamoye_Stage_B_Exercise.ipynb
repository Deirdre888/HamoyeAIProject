```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\Desktop\Data_Set\energydata_complete.csv')
df.head(5)
```

```python
#fitting a linear model and calculating the R-Squared
from sklearn.linear_model import LinearRegression

#initiating linear regression model
linear_model = LinearRegression()

#defining predictor and response variable
x = df.T2.values.reshape(-1,1)
y = df.T6.values.reshape(-1,1)

#fitting regression model
linear_model.fit(x,y)

#calculating r2 of regression model
r2 = linear_model.score(x,y)
round(r2,2)
```

```python
# Normalising our dataset to common scale and dropping date and lights column
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = df.drop(columns = ['date','lights'])
normalised_df = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
normalised_df.head(5)
```

```python
from sklearn.model_selection import train_test_split
X = normalised_df.drop(columns = 'Appliances')
Y = normalised_df['Appliances']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=42)
x_test
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test) # independent variable
predicted_values
```

```python
#MEAN ABSOLUTE ERROR 
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae,2)
```

```python
#RESIDUAL SUM OF SQUARES
import statsmodels.api as sm
x_constant = sm.add_constant(predicted_values)
model = sm.OLS(y_test, x_constant).fit()
model.ssr
```

```python
#ROOT MEAN SQUARE ERROR
import numpy as np
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#COEFFICIENT OF DETERMINATION
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,predicted_values)
round(r2_score,2)
```

```python
linear_model.coef_
```

```python
def get_weights_df(model, feat,col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
#OBTAINING THE FEATURE WEIGHT OF LINEAR MODEL
linear_model_weights = get_weights_df(linear_model,x_train,"Linear_Model_Weight")
linear_model_weights
```

```python
#TRAINING A RIDGE REGRESSION MODEL WITH AN ALPHA VALUE OF 0.4
from sklearn.linear_model import Ridge
rr = Ridge(alpha = 0.4)
rr.fit(x_train, y_train)
predicted_values_ridge = rr.predict(x_test)

rmse_ridge = np.sqrt(mean_squared_error(y_test, predicted_values_ridge))
round(rmse_ridge, 3)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha = 0.001)
lasso.fit(x_train, y_train)
predicted_values_lasso = lasso.predict(x_test)

lasso_weight_df = get_weights_df(lasso, x_train, 'Lasso_weight')
lasso_weight_df
# answer: just 4 non zero weights.
```

```python
#New RMSE with lasso Regression
rmse = np.sqrt(mean_squared_error(y_test, predicted_values_lasso))
round(rmse, 3)
```
